## Pull Data from NYT API
___

```python
articles = pd.DataFrame({'dummy': ['dummy', 'dummy'], 'dummy': ['dummy', 'dummy']})
if articles.shape[0] % 12_000  == 0:
    print('True')
else:
    print('False')
```

```python
# Function Credit:
# https://practicaldatascience.co.uk/data-science/how-to-add-days-and-subtract-days-from-dates-in-pandas

def add_days_to_date(date, days):
    """Add days to a date and return the date.
    
    Args: 
        date (string): Date string in YYYY-MM-DD format. 
        days (int): Number of days to add to date
    
    Returns: 
        date (date): Date in YYYY-MM-DD with X days added. 
    """
    
    added_date = pd.to_datetime(date) + timedelta(days=days)
    added_date = added_date.strftime("%Y-%m-%d")

    return added_date
```

```python
begin_date = "20011007"
```

```python
apikey = os.environ.get('apikey')
```

```python
# Create empty df
articles = pd.DataFrame({'dummy': ['dummy', 'dummy'], 'dummy': ['dummy', 'dummy']})

#Filter
query = "Afghanistan"

end_date = '2021-8-30'

while pd.to_datetime(begin_date) <= pd.to_datetime(end_date):
    if articles.shape[0] % 12_000  != 0:
        query_with_date = f"{query}&facet_fields=source&facet=true&sort=oldest&begin_date={begin_date}&end_date=20210830"
        url = f"https://api.nytimes.com/svc/search/v2/articlesearch.json?q={query_with_date}&api-key={apikey}"

        print(f'begin_date: {begin_date}, size: {articles.shape[0]}')
        res = requests.get(url)
        if res.status_code == 200:
    #         print(f'status code{res.status_code}')
            if int(articles.shape[0]) == 2:
    #             if this is the first time, overwrite the existing df
                articles = pd.DataFrame(res.json()['response']['docs'])
                begin_date = add_days_to_date(begin_date, 1)
                print(articles['headline'].loc[articles.shape[0]-1]['main'])
                time.sleep(2)

            else:
    #             otherwise merge the new data with the existing df
                articles2 = pd.DataFrame(res.json()['response']['docs'])
                articles = pd.concat([articles, 
                           articles2], ignore_index=True)
                begin_date = re.search(r'[0-9]{4}[\-][0-9]{2}[\-][0-9]{2}', articles['pub_date'].values[-1]).group(0).replace('-', '')
                begin_date = add_days_to_date(begin_date, 1)
                print(articles['headline'].loc[articles.shape[0]-1]['main'])
                time.sleep(2)
        else:
            print(f'a problem occured: {res.status_code}')
            break
            time.sleep(2)
    else:
        print('sleeping', f'shape= {articles.shape[0]}')
        
        #pull one more post here
        articles2 = pd.DataFrame(res.json()['response']['docs'])
        articles = pd.concat([articles, 
                   articles2], ignore_index=True)
        begin_date = re.search(r'[0-9]{4}[\-][0-9]{2}[\-][0-9]{2}', articles['pub_date'].values[-1]).group(0).replace('-', '')
        begin_date = add_days_to_date(begin_date, 1)
        print(articles['headline'].loc[articles.shape[0]-1]['main'])
        time.sleep(2)
        
        articles.to_csv(f'datasets/articles{begin_date}.csv')
        for x in range(1, 24):
            time.sleep(3600)
```
